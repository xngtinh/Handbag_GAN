```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
import os
import requests as rq
driver = webdriver.Chrome('D:/NCKH/Soft/chromedriver.exe')
```

```python
# Chọn trang cần Crawl
driver.get('https://fado.vn/us/s/search?rh=n:172282,n:541966,n:3011391011,n:172470&keywords=satchel&i=computers&sort=relevancerank')
```

```python
# Tạo danh sách rỗng và sô page cần crawl
links = []
count = 0
# begin = 1
# end = 61
```

```python
driver.execute_script("window.scrollTo(0, 2000)") 
time.sleep(3)
all = driver.find_elements_by_class_name('product-card')
# time.sleep(50)
for element in all:
        product = element.find_element_by_class_name('product-card__img') 
        links.append(product.get_property('src'))
```

```python
len(links)
```

```python
# Bắt đầu crawl trong từng page và lưu links
while True:
    scheight = .1
    while scheight < 9.9:
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight/%s);" % scheight)
        scheight += .01
    all = driver.find_elements_by_class_name('product-card')
    # time.sleep(50)
    for element in all:
            product = element.find_element_by_class_name('product-card__img') 
            links.append(product.get_property('src'))
    print(count)
#     if count == 1:
#         break
    count += 1
    

    next = driver.find_element_by_class_name('item-next')
    a = next.find_element_by_class_name('pagination-nav__page-btn')
    url = a.get_property('href')
    driver.get(url)
    
```

```python
len(links)
```

```python
# Xóa những links giống nhau
from collections import OrderedDict
links = list( set(links) )
links = list( OrderedDict.fromkeys(links) )
len(links)
```

```python
# Lưu dataset
if not os.path.exists('D://NCKH/Datasets/satchel/satchel_fado'):
    os.mkdir('D://NCKH/Datasets/satchel/satchel_fado')
i = 0

for index, img_link in enumerate(links):
    img_data = rq.get(img_link).content
    with open("D://NCKH/Datasets/satchel/satchel_fado/"+'satchel_'+str(index+1)+'.jpg', 'wb+') as f:
        f.write(img_data)
    i += 1
```
